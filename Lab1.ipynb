**Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.**

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load("/content/drive/MyDrive/speech.mp3")

plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

signal_length = len(y) / sr
magnitude_range = max(abs(y)) - min(abs(y))

print(f"Signal Length: {signal_length} seconds")
print(f"Magnitude Range: {magnitude_range}")
```

**A3. Take a small segment of the signal and play it.**

```python
from IPython.display import Audio

start_time = 0.5
end_time = 2.5

segment = y[int(start_time * sr):int(end_time * sr)]

Audio(segment, rate=sr)
```

**A4. Play around with your recorded speech signal for various segments. Understand the nature of
the signal. Also observe with abruptly segmented speech, how the perception of that speech is
affected.**

```python
from IPython.display import Audio

segment_1_start = 0.0
segment_1_end = 0.9

segment_2_start = 1.0
segment_2_end = 1.9

abrupt_segment_start = 2.0
abrupt_segment_end = 2.9

segment_1 = y[int(segment_1_start * sr):int(segment_1_end * sr)]
segment_2 = y[int(segment_2_start * sr):int(segment_2_end * sr)]
abrupt_segment = y[int(abrupt_segment_start * sr):int(abrupt_segment_end * sr)]

print("Playing Segment 1:")
display(Audio(segment_1, rate=sr))

print("Playing Segment 2:")
display(Audio(segment_2, rate=sr))

print("Playing Abruptly Segmented Speech:")
display(Audio(abrupt_segment, rate=sr))
```
